```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import rcParams
```

```python
df = pd.read_json('./group/brent-annual_json.json')
df.head(10)
```

```python
plt.plot(df['Date'], df['Price'], label = 'нефть марки brand')
title_dict = {'fontsize':20, 'fontweight':'bold', 'color':'red', 'family': 'calibri'}
plt.title('Стоимость нефти марки бред в годовом режиме', fontdict=title_dict)

fig_size = plt.gcf().get_size_inches()
sizefactor = 3 
plt.gcf().set_size_inches(sizefactor * fig_size)
plt.xlabel('Года')
plt.ylabel('Цена нефти $')
plt.grid()
plt.legend(loc='lower right', 
         fontsize = 25,
          ncol = 2,    #  количество столбцов
          facecolor = 'oldlace',    #  цвет области
          edgecolor = 'r',    #  цвет крайней линии
          title = 'Легенда',    #  заголовок
          title_fontsize = '20')   #  размер шрифта заголовка)
plt.show()
```
