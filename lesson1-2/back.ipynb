```python
target_word = 'анна'
target_chapter = 4
import math


def read_data():
    handle = open('war_peace_processed.txt', 'rt', encoding="utf-8")
    res = handle.read()
    handle.close()
    return res.split('\n')


def draw_dict(my_data):
    words = {}
    for word in my_data:
        value = words.get(word, 0)
        words[word] = value + 1
    return words


def get_chapter(my_data):
    return [chapter[0] for chapter in enumerate(my_data) if chapter[1] == '[new chapter]']


def get_info(my_data, word):
    chapters = get_chapter(my_data);
    words = {}
    previous_index = 0;
    count_chapter_with_word = 0
    last_index = len(my_data)
    chapters.append(last_index)

    for chapter in chapters:
        chapter_list = my_data[previous_index:chapter]
        previous_index = chapter
        words[chapter] = draw_dict(chapter_list)
        count_target = words[chapter].get(word, 0)
        if count_target > 0:
            count_chapter_with_word += 1

    return count_chapter_with_word, len(chapters)


def get_info_by_chapter(number_chapter, my_data):
    chapters = get_chapter(my_data)
    words = {}
    previous_index = 0
    last_index = len(my_data)
    chapters.append(last_index)

    index = 0

    for chapter in chapters:
        chapter_list = my_data[previous_index:chapter]
        previous_index = chapter
        words[chapter] = draw_dict(chapter_list)
        if index == number_chapter:
            return words[chapter]
        index += 1
    return False


def get_number(word, inform, df):
    count_target = inform.get(word, 0)
    count_words = -1

    for item in inform:
        count_words += inform.get(item, 0)

    tf = count_target
    chapter_size = count_words
    return math.log(1 + tf / chapter_size) * math.log(1 / df)


def get_result(inform, my_data):
    arr = {}
    for word in inform:
        count_chapter_with_word, count_chapters = get_info(my_data, word)
        df = count_chapter_with_word / count_chapters

        arr[word] = get_number(word, inform, df)
        
    lst = list(arr.items())
    sorted_arr = sorted(lst, key=lambda i: i[1], reverse=True)
    return sorted_arr


data = read_data()
info = get_info_by_chapter(target_chapter, data)
s_arr = get_result(info, data)
arres = s_arr[:3]
res = ''

index=0
for item in arres:
    res += item[0]
    if index != 2:
        res += ' '

print(res)

[('павловна', 0.014125754682156953), ('анна', 0.011031063403921838), ('тетушку', 0.00969211136471989), ('милее', 0.00969211136471989), ('прядильной', 0.00969211136471989), ('княгиня', 0.008686259279433266), ('исповедь', 0.008385518743031862), ('приехала', 0.00832154763737128), ('болконская', 0.007621211055625422), ('тетушки', 0.007621211055625422), ('хозяйки', 0.007621211055625422), ('веретена', 0.007621211055625422), ('здоровье', 0.007282287086581739), ('тетушкой', 0.007078926121343834), ('посланника', 0.006658297247849924), ('толстый', 0.006524121758777686), ('тетушка', 0.006314618433937395), ('павловну', 0.006314618433937395), ('элен', 0.0059269218502031795), ('маленькая', 0.0059269218502031795), ('мсье', 0.005772333499655808), ('хозяин', 0.005772333499655808), ('обратилась', 0.005549536598658905), ('она', 0.005435748396358091), ('молодая', 0.005351704626161897), ('павловне', 0.005351704626161897), ('делала', 0.005172043536574128), ('слишком', 0.0050881060766819535), ('хороши', 0.005008025812249368), ('анну', 0.005008025812249368), ('павловны', 0.004857144131995415), ('петербурга', 0.004857144131995415), ('красавица', 0.004857144131995415), ('помощь', 0.004857144131995415), ('наполняться', 0.004848339403566276), ('разнородные', 0.004848339403566276), ('возрастам', 0.004848339403566276), ('характерам', 0.004848339403566276), ('л', 0.004848339403566276), ('отцветающая', 0.004848339403566276), ('дряхлый', 0.004848339403566276), ('заехавшая', 0.004848339403566276), ('шифре', 0.004848339403566276), ('бальном', 0.004848339403566276), ('обворожительная', 0.004848339403566276), ('выезжавшая', 0.004848339403566276), ('ездившая', 0.004848339403566276), ('приезжавшим', 0.004848339403566276), ('подводила', 0.004848339403566276), ('старушке', 0.004848339403566276), ('бантах', 0.004848339403566276), ('выплывшей', 0.004848339403566276), ('отходила', 0.004848339403566276), ('совершали', 0.004848339403566276), ('интересной', 0.004848339403566276), ('приветствиями', 0.004848339403566276), ('молчаливо', 0.004848339403566276), ('подходившие', 0.004848339403566276), ('исполненной', 0.004848339403566276), ('уголку', 0.004848339403566276), ('трельяжем', 0.004848339403566276), ('мешке', 0.004848339403566276), ('черневшимися', 0.004848339403566276), ('зубам', 0.004848339403566276), ('вытягивалась', 0.004848339403566276), ('привлекательных', 0.004848339403566276), ('короткость', 0.004848339403566276), ('полуоткрытый', 0.004848339403566276), ('живости', 0.004848339403566276), ('переносившую', 0.004848339403566276), ('светлую', 0.004848339403566276), ('улыбочку', 0.004848339403566276), ('шажками', 0.004848339403566276), ('рабочею', 0.004848339403566276), ('сумочкой', 0.004848339403566276), ('серебряного', 0.004848339403566276), ('развлечением', 0.004848339403566276), ('развела', 0.004848339403566276), ('изящное', 0.004848339403566276), ('грудей', 0.004848339403566276), ('опоясанное', 0.004848339403566276), ('лиз', 0.004848339403566276), ('становитесь', 0.004848339403566276), ('дурака', 0.004848339403566276), ('стриженою', 0.004848339403566276), ('коричневом', 0.004848339403566276), ('покрой', 0.004848339403566276), ('неповоротлив', 0.004848339403566276), ('неловки', 0.004848339403566276), ('неуклюжи', 0.004848339403566276), ('мужицкие', 0.004848339403566276), ('парни', 0.004848339403566276), ('незастенчив', 0.004848339403566276), ('удивленному', 0.004848339403566276), ('бесстрастная', 0.004848339403566276), ('занятая', 0.004848339403566276), ('навестить', 0.004848339403566276), ('предпоследнее', 0.004848339403566276), ('приглядывалась', 0.004848339403566276), ('помещала', 0.004848339403566276), ('перемещала', 0.004848339403566276), ('ничтожнее', 0.004848339403566276), ('прозвучит', 0.004848339403566276), ('труба', 0.004848339403566276), ('предстану', 0.004848339403566276), ('цитируя', 0.004848339403566276), ('полюбишь', 0.004848339403566276), ('духовником', 0.004848339403566276), ('представлю', 0.004848339403566276), ('отделавшись', 0.004848339403566276), ('приглядываться', 0.004848339403566276), ('ослабевал', 0.004848339403566276), ('мастерской', 0.004848339403566276), ('посадив', 0.004848339403566276), ('работников', 0.004848339403566276), ('прохаживается', 0.004848339403566276), ('заведению', 0.004848339403566276), ('примечает', 0.004848339403566276), ('вертятся', 0.004848339403566276), ('неподвижность', 0.004848339403566276), ('непривычный', 0.004848339403566276), ('скрипящий', 0.004848339403566276), ('сдерживает', 0.004848339403566276), ('надлежащий', 0.004848339403566276), ('замолкнувшему', 0.004848339403566276), ('перемещением', 0.004848339403566276), ('равномерную', 0.004848339403566276), ('приличную', 0.004848339403566276), ('разговорную', 0.004848339403566276), ('машину', 0.004848339403566276), ('платье', 0.004826219939714658), ('замечая', 0.004717449515411831), ('анне', 0.004587396938755458), ('одинаковые', 0.004194735224481184), ('з', 0.004194735224481184), ('вышедшая', 0.004194735224481184), ('нужной', 0.004194735224481184), ('облегчения', 0.004194735224481184), ('переходили', 0.004194735224481184), ('золотом', 0.004194735224481184), ('бархатном', 0.004194735224481184), ('коротка', 0.004194735224481184), ('открывалась', 0.004194735224481184), ('опускалась', 0.004194735224481184), ('скучающим', 0.004194735224481184), ('смотревшим', 0.004194735224481184), ('побыв', 0.004194735224481184), ('обошла', 0.004194735224481184), ('самовара', 0.004194735224481184), ('захватила', 0.004194735224481184), ('сыграйте', 0.004194735224481184), ('хозяйке', 0.004194735224481184), ('одета', 0.004194735224481184), ('кружевах', 0.004194735224481184), ('серенькое', 0.004194735224481184), ('широкою', 0.004194735224481184), ('покидает', 0.004194735224481184), ('гадкая', 0.004194735224481184), ('безумия', 0.004194735224481184), ('светлых', 0.004194735224481184), ('модный', 0.004194735224481184), ('неуклюж', 0.004194735224481184), ('здоровые', 0.004194735224481184), ('знаки', 0.004194735224481184), ('бессмысленных', 0.004194735224481184), ('замахала', 0.004194735224481184), ('больную', 0.004194735224481184), ('переглядываясь', 0.004194735224481184), ('встанет', 0.004194735224481184), ('руссо', 0.004194735224481184), ('элоиза', 0.004194735224481184), ('фрейлине', 0.004194735224481184), ('сочинение', 0.004194735224481184), ('исповеди', 0.004194735224481184), ('противоположного', 0.004194735224481184), ('оглядывала', 0.004194735224481184), ('книга', 0.004194735224481184), ('гадкие', 0.004194735224481184), ('грехи', 0.004194735224481184), ('умеющего', 0.004194735224481184), ('занятиям', 0.004194735224481184), ('говорившему', 0.004194735224481184), ('заводила', 0.004194735224481184), ('гости', 0.00395314182800539), ('гостей', 0.00395314182800539), ('пьер', 0.0038824516419133826), ('василия', 0.003865450914886101), ('маленькой', 0.003865450914886101), ('отвечала', 0.003865450914886101), ('продолжала', 0.003865450914886101), ('молодой', 0.00382120666980433), ('гостиная', 0.003812401289401771), ('звездах', 0.003812401289401771), ('иностранных', 0.003812401289401771), ('дворов', 0.003812401289401771), ('кряхтя', 0.003812401289401771), ('прошлую', 0.003812401289401771), ('причине', 0.003812401289401771), ('беременности', 0.003812401289401771), ('высоких', 0.003812401289401771), ('переводя', 0.003812401289401771), ('следила', 0.003812401289401771), ('выражениях', 0.003812401289401771), ('старушки', 0.003812401289401771), ('подойти', 0.003812401289401771), ('присутствующих', 0.003812401289401771), ('шитом', 0.003812401289401771), ('хорошенькую', 0.003812401289401771), ('будущую', 0.003812401289401771), ('развертывая', 0.003812401289401771), ('прелестная', 0.003812401289401771), ('обворожительный', 0.003812401289401771), ('вкус', 0.003812401289401771), ('тогдашней', 0.003812401289401771), ('жабо', 0.003812401289401771), ('движениях', 0.003812401289401771), ('приняв', 0.003812401289401771), ('племяннице', 0.003812401289401771), ('любезно', 0.003812401289401771), ('бедную', 0.003812401289401771), ('прислушивалась', 0.003812401289401771), ('собеседницу', 0.003812401289401771), ('страницу', 0.003812401289401771), ('прислушиваться', 0.003812401289401771), ('готовая', 0.003812401289401771), ('пускает', 0.003812401289401771), ('подходила', 0.003812401289401771), ('анны', 0.003701433190561341), ('нее', 0.003685958159135711), ('идет', 0.00362448313097996), ('орденах', 0.003541131045396093), ('известная', 0.003541131045396093), ('знакомы', 0.003541131045396093), ('приезжать', 0.003541131045396093), ('приветствия', 0.003541131045396093), ('тяжелой', 0.003541131045396093), ('чайного', 0.003541131045396093), ('группы', 0.003541131045396093), ('верхняя', 0.003541131045396093), ('особенною', 0.003541131045396093), ('похожи', 0.003541131045396093), ('оправляя', 0.003541131045396093), ('высоким', 0.003541131045396093), ('сообразить', 0.003541131045396093), ('суда', 0.003541131045396093), ('прочтя', 0.003541131045396093), ('кузине', 0.003541131045396093), ('возвратилась', 0.003541131045396093), ('высшая', 0.0033307174972128117), ('саблей', 0.0033307174972128117), ('небольшие', 0.0033307174972128117), ('обряд', 0.0033307174972128117), ('известной', 0.0033307174972128117), ('одобряя', 0.0033307174972128117), ('поспешности', 0.0033307174972128117), ('губка', 0.0033307174972128117), ('здоровья', 0.0033307174972128117), ('любезен', 0.0033307174972128117), ('ридикюль', 0.0033307174972128117), ('лентой', 0.0033307174972128117), ('спокойны', 0.0033307174972128117), ('особа', 0.0033307174972128117), ('очках', 0.0033307174972128117), ('решителен', 0.0033307174972128117), ('гостем', 0.0033307174972128117), ('выражал', 0.0033307174972128117), ('поступок', 0.0033307174972128117), ('громкий', 0.0033307174972128117), ('дурно', 0.003280804317067431), ('нею', 0.003218995019174352), ('стуча', 0.00315879711031668), ('гостям', 0.00315879711031668), ('торжественным', 0.00315879711031668), ('отходили', 0.00315879711031668), ('хорошенькая', 0.00315879711031668), ('нижнюю', 0.00315879711031668), ('недостаток', 0.00315879711031668), ('старикам', 0.00315879711031668), ('переваливаясь', 0.00315879711031668), ('ниже', 0.00315879711031668), ('панталонах', 0.00315879711031668), ('моде', 0.00315879711031668), ('кланяясь', 0.00315879711031668), ('вспомнить', 0.00315879711031668), ('праздник', 0.0030134404943343257), ('видали', 0.0030134404943343257), ('участием', 0.0030134404943343257), ('разместились', 0.0030134404943343257), ('собственно', 0.0030134404943343257), ('красотой', 0.0030134404943343257), ('мрачным', 0.0030134404943343257), ('маленькими', 0.0030134404943343257), ('окружавших', 0.0030134404943343257), ('дурной', 0.0030134404943343257), ('всё', 0.0030134404943343257), ('спор', 0.0030134404943343257), ('книгой', 0.0030134404943343257), ('конечно', 0.0030134404943343257), ('полная', 0.0030134404943343257), ('местам', 0.0030134404943343257), ('никому', 0.002990228020229893), ('понемногу', 0.002887526866311002), ('дипломат', 0.002887526866311002), ('разговаривали', 0.002887526866311002), ('усиками', 0.002887526866311002), ('фраке', 0.002887526866311002), ('посередине', 0.002887526866311002), ('следует', 0.002887526866311002), ('отчаянным', 0.002887526866311002), ('новая', 0.002887526866311002), ('вызывал', 0.002887526866311002), ('встать', 0.002887526866311002), ('каждому', 0.0027764631752372664), ('приличия', 0.0027764631752372664), ('выказывая', 0.0027764631752372664), ('свободно', 0.0027764631752372664), ('смотрите', 0.0027764631752372664), ('будьте', 0.0027764631752372664), ('находя', 0.0027764631752372664), ('старую', 0.0027764631752372664), ('вечер', 0.0027385193827858434), ('обращаясь', 0.0026919734451760027), ('обществу', 0.0026771133181277204), ('грустным', 0.0026771133181277204), ('полную', 0.0026771133181277204), ('делаются', 0.0026771133181277204), ('поговорив', 0.0026771133181277204), ('василию', 0.0026771133181277204), ('приглашая', 0.0026771133181277204), ('последнего', 0.0026771133181277204), ('пойдемте', 0.0026771133181277204), ('жили', 0.0025872404403658373), ('называла', 0.0025872404403658373), ('гостя', 0.0025872404403658373), ('величества', 0.0025872404403658373), ('казались', 0.0025872404403658373), ('шутки', 0.0025872404403658373), ('писали', 0.0025872404403658373), ('красивой', 0.0025872404403658373), ('пусть', 0.0025872404403658373), ('пункт', 0.0025872404403658373), ('зиму', 0.002505192931231589), ('обязанности', 0.002505192931231589), ('ипполит', 0.002505192931231589), ('извините', 0.002505192931231589), ('книгу', 0.002505192931231589), ('кружку', 0.002505192931231589), ('ней', 0.002502175244410301), ('знаете', 0.0024345265546073178), ('сел', 0.0024345265546073178), ('одних', 0.002429716539336801), ('работой', 0.002429716539336801), ('блестящие', 0.002429716539336801), ('быстрыми', 0.002429716539336801), ('работу', 0.002429716539336801), ('платья', 0.002429716539336801), ('некогда', 0.002429716539336801), ('ход', 0.002429716539336801), ('замуж', 0.0023598363152492344), ('слава', 0.0023598363152492344), ('мужчин', 0.0023598363152492344), ('белые', 0.0023598363152492344), ('виднелись', 0.0023598363152492344), ('диван', 0.0023598363152492344), ('покраснела', 0.0023598363152492344), ('подать', 0.0023598363152492344), ('словом', 0.0023598363152492344), ('кто', 0.0023347484454586333), ('подле', 0.002317890509291933), ('каждом', 0.0022947793830483073), ('слове', 0.0022947793830483073), ('дожидаясь', 0.0022947793830483073), ('влюблен', 0.0022947793830483073), ('бывают', 0.0022947793830483073), ('самую', 0.0022947793830483073), ('вашей', 0.0022947793830483073), ('свидание', 0.0022947793830483073), ('сказала', 0.002254899667584633), ('медленно', 0.0022339226872259113), ('разу', 0.0022339226872259113), ('вскоре', 0.0022339226872259113), ('княгини', 0.0022339226872259113), ('дам', 0.0021767566086700334), ('будете', 0.0021767566086700334), ('видом', 0.002172817815748462), ('думал', 0.002146879210889204), ('самая', 0.002122858996152176), ('имени', 0.002122858996152176), ('окна', 0.002122858996152176), ('зубы', 0.002122858996152176), ('вставая', 0.002122858996152176), ('каком', 0.0020718762283290088), ('отцом', 0.0020718762283290088), ('губы', 0.0020718762283290088), ('пришли', 0.0020718762283290088), ('любят', 0.0020718762283290088), ('мнения', 0.0020718762283290088), ('торопливо', 0.0020235091390426296), ('весело', 0.0020058933603993518), ('жена', 0.001977502380169821), ('маленький', 0.001977502380169821), ('сих', 0.001977502380169821), ('вошел', 0.0019430537182905288), ('разговор', 0.0019430537182905288), ('десять', 0.001933636261280746), ('муж', 0.001933636261280746), ('генералу', 0.001933636261280746), ('подошла', 0.001933636261280746), ('значительно', 0.001933636261280746), ('серьезно', 0.0018917204047370828), ('женщин', 0.0018917204047370828), ('молодым', 0.0018917204047370828), ('ответа', 0.0018917204047370828), ('словах', 0.0018917204047370828), ('жить', 0.0018917204047370828), ('ее', 0.00188959106947122), ('нынче', 0.0018822415189109242), ('человек', 0.0018669748719411941), ('или', 0.0018597182081766849), ('чувством', 0.001851588752146498), ('видите', 0.001851588752146498), ('показать', 0.001851588752146498), ('василий', 0.001851588752146498), ('мадемуазель', 0.001851588752146498), ('очки', 0.001851588752146498), ('и', 0.0018510978682366069), ('начала', 0.0018130955908593478), ('свет', 0.0018130955908593478), ('стола', 0.0018130955908593478), ('рот', 0.0018130955908593478), ('стол', 0.0018130955908593478), ('села', 0.0018130955908593478), ('неловко', 0.0018130955908593478), ('говорила', 0.0017945506057916347), ('скоро', 0.0017945506057916347), ('смерть', 0.0017761123602517102), ('почему', 0.0017761123602517102), ('никакого', 0.0017761123602517102), ('богу', 0.0017405250610727627), ('руке', 0.0017405250610727627), ('тоном', 0.0017405250610727627), ('дочери', 0.0017405250610727627), ('ваш', 0.0017405250610727627), ('гораздо', 0.0017405250610727627), ('звук', 0.0017405250610727627), ('лучше', 0.0017382773315421625), ('людям', 0.0017062321361641433), ('узнав', 0.0017062321361641433), ('первую', 0.0017062321361641433), ('хочу', 0.0017062321361641433), ('совершенно', 0.0016836353985418661), ('дочь', 0.0016731427232349449), ('женщина', 0.0016731427232349449), ('мать', 0.0016731427232349449), ('беспрестанно', 0.0016731427232349449), ('вообще', 0.0016731427232349449), ('самые', 0.0016411752039632163), ('скажите', 0.0016411752039632163), ('война', 0.0016411752039632163), ('несмотря', 0.001631603624974328), ('эта', 0.0016305328814668746), ('вы', 0.0016121118905596365), ('вошла', 0.001610255991089475), ('моей', 0.001610255991089475), ('немного', 0.001610255991089475), ('лицу', 0.00158031850814082), ('руках', 0.00158031850814082), ('одним', 0.00158031850814082), ('чуть', 0.0015513023262013327), ('сами', 0.0015513023262013327), ('иногда', 0.0014958185879808616), ('хуже', 0.0014958185879808616), ('человека', 0.001488111323980994), ('легко', 0.0014692548170670847), ('столько', 0.0014692548170670847), ('вполне', 0.001443418913042301), ('мной', 0.001443418913042301), ('всем', 0.0014326940339950427), ('сын', 0.0014182720492439177), ('правда', 0.0014182720492439177), ('--', 0.0013959522235953785), ('смотреть', 0.001393778425172297), ('своею', 0.001393778425172297), ('другим', 0.0013699049599575383), ('улыбаясь', 0.0013625318807517625), ('вечера', 0.00132389820108473), ('тихо', 0.00132389820108473), ('вместе', 0.001317648458048252), ('ехать', 0.0013017100830708477), ('ко', 0.0013017100830708477), ('какое', 0.0013017100830708477), ('весьма', 0.001280032082195655), ('положение', 0.001280032082195655), ('молодого', 0.001280032082195655), ('свое', 0.0012738089310910777), ('большой', 0.001258841268883803), ('гостиной', 0.001258841268883803), ('стали', 0.0012381162256519917), ('бывает', 0.0012381162256519917), ('встал', 0.0012381162256519917), ('комнаты', 0.0012178369170103148), ('каждый', 0.001197984573061407), ('желая', 0.001197984573061407), ('генерал', 0.0011785415851023754), ('мысль', 0.0011785415851023754), ('руками', 0.0011594914117742567), ('особенно', 0.001148094710079885), ('знать', 0.0011408184945055292), ('понять', 0.0011408184945055292), ('вас', 0.0010882479471852875), ('остановился', 0.0010869208819876714), ('пор', 0.0010869208819876714), ('начал', 0.0010696185340123732), ('князя', 0.0010302429183851947), ('эту', 0.0010302429183851947), ('около', 0.0010195385441498538), ('стороны', 0.0010195385441498538), ('тотчас', 0.0010034193224375126), ('дела', 0.0010034193224375126), ('головой', 0.0009875710248781252), ('продолжал', 0.0009875710248781252), ('времени', 0.0009719846946791866), ('зачем', 0.0009719846946791866), ('делать', 0.0009566518120043839), ('приехал', 0.0009415642660053167), ('могу', 0.0009267143290557291), ('наконец', 0.0009120946329833372), ('уж', 0.0008976981471162417), ('сделал', 0.0008835181579801784), ('одной', 0.0008695482504998513), ('совсем', 0.0008695482504998513), ('тем', 0.0008630669415463993), ('не', 0.0008586079986842232), ('своем', 0.0008557822905725786), ('тогда', 0.0008422144088957704), ('дома', 0.0008422144088957704), ('весь', 0.0008156506379819936), ('свой', 0.0008156506379819936), ('князю', 0.0008156506379819936), ('ли', 0.0008145777250667774), ('граф', 0.000802644204078766), ('их', 0.0007803761518318488), ('по', 0.0007762074826089472), ('со', 0.0007643076874064397), ('всех', 0.0007530845332026992), ('видел', 0.0007523415311338869), ('глядя', 0.0007523415311338869), ('с', 0.0007425585154930532), ('своей', 0.0007315239968094908), ('говорили', 0.0007281619625215789), ('улыбкой', 0.0007163007808724473), ('люди', 0.0007045869469082582), ('чтоб', 0.0007045869469082582), ('на', 0.0007038036414337501), ('много', 0.0006930168445970978), ('я', 0.0006892142218007637), ('до', 0.0006890925638549291), ('другой', 0.0006833933472981852), ('тот', 0.0006815869894056831), ('тех', 0.000670294021999639), ('свою', 0.000667619073691377), ('тут', 0.0006591347023165696), ('этих', 0.0006481059039857569), ('у', 0.00062780393457593), ('говорить', 0.0006264279031105638), ('в', 0.0006261859242752088), ('мой', 0.0006157729704456027), ('минуту', 0.0006157729704456027), ('при', 0.000605809775798298), ('казалось', 0.0006052370897987119), ('ей', 0.0006052370897987119), ('кроме', 0.000594817630104851), ('котором', 0.000594817630104851), ('стал', 0.0005845120465669008), ('говорил', 0.0005652979755911897), ('потом', 0.0005642327379252237), ('хорошо', 0.0005346087883154077), ('о', 0.0005234157193957419), ('несколько', 0.0005153642120368283), ('есть', 0.00048721431542043817), ('всегда', 0.0004689040020815281), ('даже', 0.0004689040020815281), ('которые', 0.00045961136427824035), ('князь', 0.00045094247791170307), ('через', 0.00045094247791170307), ('была', 0.000446699123878668), ('ни', 0.0004455963205665242), ('этого', 0.0004455963205665242), ('которое', 0.000442088408090311), ('надо', 0.000442088408090311), ('к', 0.00043150868049177967), ('глаза', 0.0004160143549272821), ('еще', 0.0003943920070743022), ('что', 0.0003924641856595681), ('теперь', 0.00039055548486750655), ('они', 0.00039055548486750655), ('было', 0.0003752958671455747), ('себе', 0.00037409849838361883), ('меня', 0.00036593436506476287), ('своим', 0.0003578403110077925), ('опять', 0.0003578403110077925), ('как', 0.00035338318358048564), ('чтобы', 0.000350304531439848), ('будто', 0.0003498151433524214), ('да', 0.0003498151433524214), ('за', 0.0003423975932056157), ('ним', 0.0003418576994380834), ('после', 0.0003418576994380834), ('все', 0.00034012804235280847), ('очень', 0.0003183805155940956), ('но', 0.0003153687830189535), ('он', 0.0003153687830189535), ('где', 0.0003106829095176845), ('этот', 0.00028796008692022566), ('мне', 0.00027311014997063803), ('для', 0.0002722927907851903), ('быть', 0.0002512647364881932), ('ему', 0.00023446513952609508), ('это', 0.00022547503114169121), ('сказал', 0.00022202275312485265), ('от', 0.00020563184939801588), ('который', 0.00020217811148748758), ('него', 0.00020217811148748758), ('же', 0.0001703039989472331), ('того', 0.00016204645889690257), ('когда', 0.000155520816881481), ('из', 0.00013715245207721697), ('его', 0.00013541230594719248), ('так', 0.00010191750665399528), ('был', 0.00010191750665399528), ('то', 6.86085428322897e-05), ('а', 6.269660178735637e-05), ('только', 5.6821495505132595e-05), ('[new chapter]', 5.530523231478545e-06)]


# [('павловна', 0.014125754682156953), ('анна', 0.011031063403921838), ('тетушку', 0.00969211136471989), ('милее', 0.00969211136471989), ('прядильной', 0.00969211136471989)]
```
