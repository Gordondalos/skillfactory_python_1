```python
from numpy import sqrt
import random


def avg_of_random_numbers(n):
    sum = 0
    for i in range(n):
        sum+=random.random()
    return sum/n


def get_normal_random(n):
    res = avg_of_random_numbers(n)
    result = (res-0.5)/(1/sqrt(12 * n))
    return result
    

def estimate_outliers(n, m):
    result = []
    for i in range(m):
        temp = get_normal_random(n)
        result.append(temp)
    new = []
    for i in result:
        if i > 3 or i < -3:
            new.append(i)
    return(len(new))
```

```python
res = estimate_outliers(10000,1000)
print(res)
```
