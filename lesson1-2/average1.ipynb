```python
import random
from numpy import sqrt

def avg_of_random_numbers(n):
    sum = 0
    for i in range(n):
        sum+=random.random()
    return sum/n


def get_normal_random(n):
    res = avg_of_random_numbers(n)
    result = (res-0.5)/(1/sqrt(12 * n))
    return result
    

def calc_mean_square(n, m):
    sum = 0
    for i in range(m):
        sum += get_normal_random(n) ** 2
    return sum / m

def calc_mean(n, m):
    sum = 0
    for i in range(m):
        sum += get_normal_random(n)
    return sum / m


def calc_var(n, m):
    return calc_mean_square(n, m) - calc_mean(n, m) ** 2
    
 
    
```
